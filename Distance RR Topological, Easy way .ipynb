```python
#lineSkip = 0
SD_matrix = np.array([])


for line in setofnormalAfRandom :
    
    
    line = line.strip()
    a,b = line.split(",")
    #b = b[1:len(b)-1]
    #print(lineSkip)
    print(line)
    
    FileName = "/Users/Esteban/Desktop/MSRI/training2017/"
    #if(lineSkip != -1):
    #print(lineSkip)
    sig_Total, fields_Total = wfdb.rdsamp( FileName + a, channels = [0])

    xqrs_Total = processing.XQRS(sig =sig_Total[:,0], fs = fields_Total["fs"]) 
    
    xqrs_Total.detect()
    
    Time_Total    = xqrs_Total.qrs_inds
        
    D = np.zeros( (len(Time_Total)**2) )
    D = D.reshape(len(Time_Total), len(Time_Total))
    for i in range (len(Time_Total)):
        for j in range (len(Time_Total)):
            D[i,j] = abs( Time_Total[j] - Time_Total[i] )
                
    Diagrams = ripser(D,distance_matrix= True, maxdim = 0)
        
    points = Diagrams['dgms']

    points = points[0]

    death = points[:,1]
        
    death = death[0:len(death)-2]
        
    SD_death= np.std(death)
        
    SD_matrix = np.append(SD_matrix,SD_death)
     
        

   # lineSkip+= 1
```
